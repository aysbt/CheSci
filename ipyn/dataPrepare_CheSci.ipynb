```python
def data_tag(df, isCherenkov=True):
    new_df = df[(df.numPE > 5)]

    #create a label 1 for Cherenkov 0 for Scintilator
    if isCherenkov:
        new_df['label'] = 1
    else:
        new_df['label'] = 0

    final_df = new_df
    #print( final_df)
    return  final_df
```
